## Mixing data with Proportions

```python

def mix(data1,data2,a,b):
    data1['label'] = 0
    data2['label'] = 1
    
    df1 = data1.sample(frac=a)
    df2 = data2.sample(frac=b)
    
    
    df =  pd.concat([df1, df2], axis = 0)
    
    df_final = df.sample(frac=1)
    
    df_final1 = df_final.reset_index()
    df_final2 = df_final1.drop('index', axis = 1)
    df_final3 = df_final2.dropna()
    
    return df_final3
```

## Normalization of the Data Set

```python
def normalization(data):
    #number_column = len(data.columns) - 1
    
    
    
    for i in range(0, len(data.columns) - 1):
        column_name = data.columns[i]
        column_values = data[column_name]
        # Find max and min values in the column
        max_val = column_values.max()
        min_val = column_values.min()
        # Normalize each value in the column
        normalized_values = [(val - min_val) / (max_val - min_val) for val in column_values]

        # Update the column with normalized values
        data[column_name] = normalized_values
    return data     
```

## Train Test Split

```python
def generate_sample(signal_frac, train_size, signal, background):
    
   
    ### split each into training/testing
        #background
    nb = len(background)
    bg_test_n = int(nb - np.ceil(train_size * nb))
    bg_test_vals = np.random.choice(nb, bg_test_n, replace=False)
    bg_train = background.drop(bg_test_vals)
    bg_test = background.drop(bg_train.index)
        #signal
    ns = len(signal)
    sg_test_n = int(ns - np.ceil(train_size * ns))
    sg_test_vals = np.random.choice(ns, sg_test_n, replace=False)
    sg_train = signal.drop(sg_test_vals)
    sg_test = signal.drop(sg_train.index)
   
    ### create xtrain, ytrain
    bg_train['target'], sg_train['target'] = np.zeros(len(bg_train)), np.zeros(len(sg_train)) + 1
    X_train = pd.concat([bg_train, sg_train])
    X_train = shuffle(X_train)
    X_train = X_train.reset_index(drop=True)
    y_train = X_train['target']
    X_train = X_train.drop(['target'], axis=1)
   
    ### create xtest, ytest
    bg_test['target'], sg_test['target'] = np.zeros(len(bg_test)), np.zeros(len(sg_test)) + 1
    X_test = pd.concat([bg_test, sg_test])
    X_test = shuffle(X_test)
    X_test = X_test.reset_index(drop=True)
    y_test = X_test['target']
    X_test = X_test.drop(['target'], axis=1)
   
    ###remove target from sg, bg
    sg_test, bg_test = sg_test.drop(['target'], axis=1), bg_test.drop(['target'], axis=1)
    sg_test, bg_test = sg_test.reset_index(drop=True), bg_test.reset_index(drop=True)
   
    return X_train, np.array(y_train), X_test, np.array(y_test), sg_test, bg_test
```

### Multi Layered Perceptron Function

```python
def MLP2(signal_frac, train_size, signal, background):
   
    x_train, y_train, x_test, y_test, sg_test, bg_test = generate_sample(signal_frac, train_size, signal, background)
    
    
    
    model_mlp = MLPClassifier(random_state=42)
    model_mlp.fit(x_train, y_train)
    
    #y_pred = model_mlp.predict(x_test)
    
    return x_train, y_train, x_test, y_test, sg_test, bg_test, model_mlp
```

```python
def normalization1(data1, data2):
    #number_column = len(data.columns) - 1
    min_val = min(min(data1), min(data2))
    max_val = max(max(data1), max(data2))
    # Normalize each value in the column
    val1 = (data1 - min_val) / (max_val - min_val)
    val2 = (data2 - min_val) / (max_val - min_val)
    return val1, val2  
```

```python
def output(model, data):   
    c, i, a = model.coefs_, model.intercepts_, model.activation
    output = data

    for x in range(len(c)):
        output = np.dot(output, c[x]) + i[x]
        if x != (len(c) - 1):
            output = np.maximum(0.0, output)
        
    return output
```

```python
def Curve2(score, y, n_thresholds = 100, correct_t = 0.6): #again take correct output y, produced score
    thresholds = np.linspace(0, 1, n_thresholds + 1) #from number of thresholds create array with values 0, 0.1, 0.2 etc
    FPR = [] #initialize lists for fpr, tpr, and the zscore
    TPR = []
    zscore = []
    P = sum(y) #since y is 0 and 1 only, sum(y) counts the number of 1s ie positives
    N = len(y) - P #this counts the number of 0s

    for t in thresholds: #for each value of threshold we calculate the FPR and TPR and get the x and y coords for plotting
        fp, tp, fn, tn = 0, 0, 0, 0 #same method as the previous function
        for i in range(len(score)):
            if score[i] >= t:
                if y[i] == 1:
                    tp += 1
                elif y[i] == 0:
                    fp += 1
            if score[i] < t:
                if y[i] == 0:
                    tn += 1
                elif y[i] == 1:
                    fn += 1
        FPR.append(fp/N) #just calculate these two values also and we get the x and y coords, and add it to the initialized lists
        TPR.append(tp/P)
        z = tp / np.sqrt(tp + fp) #here from this we can also calculate the approx zscore and add it to the list
        zscore.append(z)
        # I dont remember what the variable correct_t does so you can just ignore it
        
        a = np.array(FPR)
        b = np.array(TPR)
        c = np.array(zscore)
        d = np.array(thresholds)

    return a, b, c, d #we can then plot using these outputs
    
```

## From MLP in SKlearn

```python
def MLP(data):
    # Splitting training and splitting data wiht test size  = 0.3
    from sklearn.model_selection import train_test_split
    x = data
    #x = data.drop("label", axis=1)
    y = data["label"]
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
    
    
    from sklearn.neural_network import MLPClassifier
    model_mlp = MLPClassifier(random_state=42, solver = 'sgd')
    model_mlp.fit(x_train, y_train)
    
    
    y_pred = model_mlp.predict(x_test)
    
    return y_test, y_pred, x_test, x_train
```

```python

def accuracy_measures(y_test,y_pred):
    
    # defining confusion matrix
    cm= confusion_matrix(y_test,y_pred)
    
    
    #finding the accuracy measures explicitly using the definitions for Perceptron



    TP = cm[1,1]   #known to be in signal, predicted to be in signal
    TN = cm[0,0]   #known to be in bg, predicted to be in bg
    FP = cm[0,1]  #known to be in bg, predicted to be in signal
    FN = cm[1,0]  #known to be in signal, predicted to be in bg

    Sensitivity = TP/(TP+FN)   #also known as recall
    Specificity = TN/(TN+FP)
    Precision = TP/(FP+TP)
    
    import statistics
    F1_Score = statistics.harmonic_mean([Precision, Sensitivity])
    Balanced_Accuracy = 0.5*(Sensitivity + Specificity)
    
    z_score = np.sqrt(2*(TP+TN)*(np.log((TP+TN)/TN))-2*(TP))
    z_score_approx = TP/np.sqrt(TP+TN)
    
    return z_score, z_score_approx

    
```

```python
def ROC_Curve(y_test, y_pred):
    #Defining Metric
    fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred)

    #Calculating the AUC (Area Under the Curve)
    auc = metrics.roc_auc_score(y_test, y_pred)
    
    fig = plt.plot(fpr,tpr,label="AUC="+str(auc))
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.title('Perceptron ROC and AUC')
    plt.legend(loc = 4)

    return fig
```

```python
from scipy.stats import ks_2samp
from sklearn.metrics import roc_auc_score

def evaluate_ks_and_roc_auc(y_real, y_proba):
    # Unite both visions to be able to filter
    df = pd.DataFrame()
    df['real'] = y_real
    df['proba'] = y_proba
    
    # Recover each class
    class0 = df[df['real'] == 0]
    class1 = df[df['real'] == 1]
    
    ks = ks_2samp(class0['proba'], class1['proba'])
    roc_auc = roc_auc_score(df['real'] , df['proba'])
    
    print(f"KS: {ks.statistic:.4f} (p-value: {ks.pvalue:.3e})")
    print(f"ROC AUC: {roc_auc:.4f}")
    
    return ks.statistic, roc_auc
```
