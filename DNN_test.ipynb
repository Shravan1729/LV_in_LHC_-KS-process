```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the benchmark summary once
summary_df = pd.read_csv("BP.csv", index_col="run_name")

def pT_dist(n):
    run_name = f"run_{int(n):02d}"
    
    # Load KS CSV for this run
    ks = pd.read_csv(f"KS/{run_name}.csv")
    ks['label'] = 'ks'
    
    mu1 = ks['pt_mu1']
    mu2 = ks['pt_mu2']

    # Extract MW and MN from the summary table
    MW = summary_df.loc[run_name, 'mass_MWp_GeV'] / 1000
    MN = summary_df.loc[run_name, 'mass_MNR_GeV'] / 1000

    # Plot
    plt.figure(figsize=(8, 5))
    plt.hist(mu1, bins=100, alpha=0.5, label='pt_mu1')
    plt.hist(mu2, bins=100, alpha=0.5, label='pt_mu2')
    plt.title(f'benchmark{n}(MW_{MW:.1f}TeV_MN_{MN:.1f}TeV)')
    plt.legend()
    plt.tight_layout()
    plt.savefig(f'KS/Plots_KS/benchmark{n}.png')
    plt.show()
    plt.close()
```

## Simple Classifier

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score, roc_curve
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Input
from tensorflow.keras import layers
```

```python
df_binary = df_sk[df_sk['label'].isin(['signal', 'ks'])].copy()
df_binary['y'] = (df_binary['label'] == 'signal').astype(int)

X = df_binary[features].values
y = df_binary['y'].values

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
# Standardize

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
#Model
model = keras.Sequential([
    keras.Input(shape=(X_train.shape[1],)),
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
##Exercise : Change the model and try to avoid overfitting

#Compilation
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

history = model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=64,
    validation_split=0.2,
    verbose=1
)
```

```python
# Predict
y_pred_prob = model.predict(X_test).ravel()
y_pred = (y_pred_prob > 0.5).astype(int)

# print("y_test shape:", y_test.shape)
# print("y_pred_prob shape:", y_pred_prob.shape)

# Accuracy Curve
plt.figure(figsize=(6, 4))
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.legend()
plt.show()

# Loss Curve
plt.figure(figsize=(6, 4))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.legend()
plt.show()
```

```python
fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
auc_val = roc_auc_score(y_test, y_pred_prob)

plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, label=f"AUC = {auc_val:.3f}")
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
df_scores = pd.DataFrame({
    'score': y_pred_prob,
    'true_label': y_test
})
df_scores['class'] = df_scores['true_label'].map({1: 'Signal', 0: 'KS'})  # or 'BKG'

plt.figure(figsize=(7, 4))
sns.histplot(data=df_scores, x='score', hue='class', bins=50, stat='density',
             common_norm=False, alpha=0.6, palette='Set1')
plt.xlabel("Predicted Score")
plt.show()
```
